```python
# Import Library yang dibutuhkan
import tweepy
import csv
import pandas as pd
import re
import nltk
```

```python
# Masukkan Twitter Token API
access_token = "isi dengan access token"
access_token_secret ="isi dengan access token secret"
consumer_key = "isi dengan key"
consumer_secret = "isi dengan key secret"
```

```python
#melakukan autentikasi ke API twitter
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
# Buat path file CSV kedalam variabel
csvFile = open('dataset_search_bersih.csv', 'a')

# buat variabel baru untuk membuat file csv
csvWriter = csv.writer(csvFile)

# Mulai crawling search tweet
for tweet in tweepy.Cursor(api.search,q="bitcoin",count=1000,lang="in",since="2020-03-20").items():
    tweet_clean = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())
    lower_case = tweet_clean.lower()
    print (tweet.created_at, lower_case)
    csvWriter.writerow([tweet.created_at,tweet.user.screen_name, lower_case.encode('utf-8')])
```
